```python
import pandas as pd
```

```python
df= pd.read_excel(r'C:\Users\balod\Documents\Python Programming\Pandas\Customer Call List.xlsx')
df
```

```python
#Removing Duplicates
df.drop_duplicates(inplace =True)
df
```

```python
#Removing the Non Useful Column
df.drop(columns= 'Not_Useful_Column', inplace= True)
df
```

```python
#Strip-> from both left and right sides
#lstrip-> from left side
#rstrip->from right side
#To remove the unwanted dots and underscores from column Lastname

#df["Last_Name"].str.strip(['...','_','/']) -> wont work this way

#df['Last_Name'] = df['Last_Name'].str.strip('123._/') ->will work

#Needs to be done 1 by one
df['Last_Name'] = df['Last_Name'].str.strip('...')
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('_')
df['Last_Name'] = df['Last_Name'].str.strip('/')
```

```python
df
```

```python
#Cleaning up phone number by using 
#Replace function

df['Phone_Number']= df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '')
```

```python
df
```

```python
#Converting all the values to string
df['Phone_Number']= df['Phone_Number'].apply(lambda x:str(x))
```

```python
#adding hypens
df['Phone_Number']= df['Phone_Number'].apply(lambda x: x[0:3]+ "-"+ x[3:6]+"-"+x[6:10])
```

```python
df
```

```python
#removing nan--, Na--
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('Na--','')
```

```python
df
```

```python
#Split function-> to split the contents of a single column
#Splitting the Address column

df[['Street_Address', 'State', 'Zip_Code']]= df['Address'].str.split(',',2,expand=True)
```

```python
df
```

```python
df.drop(columns= 'Address')
```

```python
#Changing the Yes, No format to a single one

df['Paying Customer']= df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df['Paying Customer']= df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes', 'Y')
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
#df.replace('NaN','', inplace= True) #wont work as they arent values, they are empty

#so alternatively we can
df.fillna('')
df
```

```python
df.replace('N/a','', inplace= True)
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace= True)
```

```python
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)

#Another way to drop null values
#or simple
#df.dropna(subset= 'Phone_Number', inplace= True)
```

```python
df
```

```python
df.reset_index(drop= True, inplace=True)
```
